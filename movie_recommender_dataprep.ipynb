## Importing required packages

```python
import pandas as pd
import numpy as np
```

## Importing csv files

```python
credits = pd.read_csv(r'datasets/tmdb_5000_credits.csv')
movies = pd.read_csv(r'datasets/tmdb_5000_movies.csv')

#credits.head()
#movies.head()
```

## Merging dataframes using 'movie_id' column

```python
movies = movies.merge(credits, on='title')
#movies.info()
```

## Removing unwanted fields

```python
#keeping: genres, id, keywords, title, overview, cast, crew
'''dropping: budget, homepage, original_language, original_title, popularity, production_companies, production_countries, release_date, revenue, runtime,
        spoken_languages, status, tagline, vote_average, vote_count, movie_id'''
movies = movies[['genres', 'id', 'keywords', 'title', 'release_date', 'overview', 'cast', 'crew']]
```

## Preprocessing the current DataFrame

```python
#checking for null values
movies.isnull().sum() #we find 3 rows with no overview	
```

```python
#removing fields with Null Values
movies.dropna(inplace=True) #removes all null values
movies.isnull().sum()
```

```python
#checking duplicate values
movies.duplicated().sum()
```

## Analyzing each Field and Converting the data in it into a proper format

```python
# packagetohelp convert a list stored as a string to a true list 
from ast import literal_eval as litEv

def fetch_year(date):
    return date.split('-')[0]

# helper function to fetch genres and keywords
def helper1(obj): # list of dictionaries stored as a string
    L = []
    for x in litEv(obj):
        L.append(x['name'])
    return L

# helper function to fetch top 3 actors
def helper2(obj):
    L = []
    for x in litEv(obj)[:3]:
        L.append(x['name'])
    return L

# helper function to fetch crew
def helper3(obj):
    for x in litEv(obj):
        if x['job'] == 'Director':
            return [x['name']]
    return []
```

```python
# analyzing the field 'genres'
movies.iloc[0].genres

# using helper1 function on the 'genres' field
movies['genres'] = movies['genres'].apply(helper1)
movies.head()
```

```python
# analyzing the field 'keywords'
movies.iloc[0].keywords

# using helper1 function on the 'keywords' field
movies['keywords'] = movies['keywords'].apply(helper1)
movies.head()
```

```python
# analyzing the fields
movies.iloc[0].cast

#using helper2 function on the cast field
movies['cast'] = movies['cast'].apply(helper2)
movies.head()
```

```python
# analyzing the fields
movies.iloc[0].crew

movies['crew'] = movies['crew'].apply(helper3)
movies.rename(columns = {'crew':'director'}, inplace = True) #renaming field 'crew' to 'director'
movies.head()
```

```python
# changing release_date into release_year
movies.iloc[0].release_date

movies['release_date'] = movies['release_date'].apply(fetch_year)
movies.rename(columns = {'release_date':'year'}, inplace = True)
movies.head()
```

```python
#changing 'overview' field into a list of strings (words)

movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies.head()
```

```python
# transforming words like 'Sam Worthington' and 'space war' into 'SamWorthington' and 'spacewar'
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['director'] = movies['director'].apply(lambda x:[i.replace(' ', '') for i in x])
movies.head()
```

```python
## Creating the 'tags' field using the fields: overview, genres, keywords, cast, director
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['director']

movies.head()
```

```python
df = movies[['id','title','year','tags']]
df['tags'] = df['tags'].apply(lambda x: " ".join(x).lower())
df.head()
```

```python
#storing df as backup
df.to_csv(r'datasets/final_df.csv')
```
